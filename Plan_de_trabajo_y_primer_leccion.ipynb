#### 3. Estructuras de control

**Explicación**:
Las estructuras de control permiten dirigir el flujo de un programa. Las más comunes son las declaraciones `if`, los bucles `for` y los bucles `while`.

```python
# Estructura if
edad = 18
if edad >= 18:
    print("Eres mayor de edad")
else:
    print("Eres menor de edad")

# Bucle for
for i in range(5):
    print(i)

# Bucle while
i = 0
while i < 5:
    print(i)
    i += 1
```

#### 4. Operadores

**Explicación**:
Python soporta operadores aritméticos (`+`, `-`, `*`, `/`, `//`, `%`, `**`), operadores de comparación (`==`, `!=`, `>`, `<`, `>=`, `<=`), y operadores lógicos (`and`, `or`, `not`).

```python
a = 10
b = 3

# Operadores aritméticos
suma = a + b
resta = a - b
producto = a * b
cociente = a / b
division_entera = a // b
modulo = a % b
potencia = a ** b

# Operadores de comparación
igual = (a == b)
diferente = (a != b)

# Operadores lógicos
y = (a > 5 and b < 5)
o = (a > 5 or b > 5)
no = not(a > 5)
```

#### 5. Funciones

Las funciones en Python se definen usando la palabra clave `def` y permiten encapsular código que puede ser reutilizado.

#### 6. Módulos y paquetes

**Explicación**:
Un módulo es un archivo que contiene código Python. Puedes usar el código de un módulo en otro archivo usando la palabra clave `import`. Un paquete es una colección de módulos.

```python
# Importar un módulo completo
import math
print(math.sqrt(16))

# Importar funciones específicas de un módulo
from math import sqrt
print(sqrt(25))
```

#### 7. Manejo de errores

**Explicación**:
El manejo de errores en Python se hace con las estructuras `try`, `except`, `else` y `finally`.

```python
try:
    resultado = 10 / 0
except ZeroDivisionError:
    print("No se puede dividir por cero")
else:
    print("La división fue exitosa")
finally:
    print("Fin del bloque try-except")
```
